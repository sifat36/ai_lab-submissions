```python
def a_star(graph, Fn, start, end):
    visited, queue = [], [start] 
    while queue:
        vertex = queue.pop(0)
        if vertex not in visited:
            visited.append(vertex)
            if vertex is end: break
            queue.extend(Fn[vertex])
    return visited

a_star(graph, Fn, 'Arad', 'Bucharest') 
```
